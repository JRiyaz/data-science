```python
import numpy as np
import pandas as pd
import seaborn as sns
```

# What is Seaborn?
- Seaborn is a Python data visualization library built on top of Matplotlib.
- It provides a higher-level interface for creating attractive and informative statistical graphics.
- Seaborn is particularly suited for working with pandas DataFrames and NumPy arrays.

# Advantages of Seaborn over Matplotlib:
- Built-in Themes: Seaborn comes with several aesthetically pleasing default themes and color palettes, making your plots look nicer with minimal effort.
- Easier Statistical Plots: Seaborn provides simple interfaces for creating complex statistical plots like regression plots, distribution plots, box plots, and heatmaps.
- Integration with Pandas: Seaborn works seamlessly with pandas DataFrames, allowing you to directly plot data from a DataFrame without the need to manually extract columns.
- Automatic Grouping and Color Coding: Seaborn automatically handles grouping and coloring of categorical data, which is very useful for working with grouped data or categorical variables.

```python
# Seaborn comes with quiet a few built in datasets

sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# Check no.of penguins per species
penguins["species"].value_counts()
```

```python
# Check no.of penguins per island
penguins["island"].value_counts()
```

```python
# Check no.of penguins species living on Dream island
penguins[penguins["island"] == "Dream"]["species"].value_counts()
```

# Scatter plot

```python
sns.set_style("ticks")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
# sns.despine()
```
